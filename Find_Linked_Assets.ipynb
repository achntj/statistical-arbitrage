```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

# ETFs

```python
etf_tickers = ['SPY', 'VOO', 'QQQ', 'XLK']
etf_df = yf.download(etf_tickers, start="2018-01-01", end="2024-01-01")
etf_df
```

```python
# Pearson Correlation

returns = etf_df['Adj Close'].pct_change().dropna()
corr_matrix = returns.corr() # this method calculates pearson correlation by default
print(corr_matrix)
```

```python
returns
```

```python
from statsmodels.tsa.stattools import coint

score, p_value, _ = coint(returns['QQQ'], returns['XLK'])
print(f'Cointegration Test p-value: {p_value}')
```

```python
data = etf_df['Adj Close']
etf1 = 'QQQ'
etf2 = 'XLK'

price_ratio = data[etf1] / data[etf2]

plt.figure(figsize=(14, 7))

plt.subplot(2, 1, 1)
plt.plot(data.index, data[etf1], label=etf1, color='b')
plt.plot(data.index, data[etf2], label=etf2, color='g')
plt.title(f"Price Movements of {etf1} and {etf2}")
plt.ylabel("Price (USD)")
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(data.index, price_ratio, label=f'{etf1}/{etf2} Ratio', color='r')
plt.title(f"Price Ratio of {etf1} to {etf2}")
plt.ylabel("Price Ratio")
plt.xlabel("Date")
plt.legend()

plt.tight_layout()
plt.show()
```
